```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
df = pd.read_csv('Sneaker10k.csv')  # Replace 'your_file.csv' with your actual CSV file path

# Group by 'NAME' and calculate the count of items sold
sold_counts = df['NAME'].value_counts()

# Calculate the cumulative percentage
cumulative_percentage = (sold_counts.cumsum() / sold_counts.sum()) * 100

# Create Pareto chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the bar chart for item counts
ax1.bar(sold_counts.index, sold_counts, color='blue', alpha=0.7, label='Count')

# Find the index where cumulative percentage crosses or reaches 80%
index_80_percent = (cumulative_percentage >= 80).idxmax()
index_80=str(index_80_percent)

# Create a secondary y-axis for cumulative percentage
ax2 = ax1.twinx()
ax2.plot(sold_counts.index, cumulative_percentage, color='red', marker='o', label='Cumulative Percentage')

# Set labels and title
ax1.set_xlabel('Item Name')
ax1.set_ylabel('Count', color='blue')
ax2.set_ylabel('Cumulative Percentage', color='red')
plt.title('Pareto Chart: Item Count by Name')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, ha='right')

# Show legend
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()

##print(f"Number of names contributing to 80% of the total volume: {str(index_80 )}")
```
